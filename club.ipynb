```python
%time
import numpy as np
from collections import Counter

class KNN:
    def __init__(self, k, metric, pb):
        self.k = k
        self.metric = metric
        self.pb = pb
    
    def fit(self, x_train, y_train):
        self.x_train = x_train
        self.y_train = y_train
        self.cov_matrix = np.cov(self.x_train, rowvar=False)
        self.inv_cov_matrix = np.linalg.inv(self.cov_matrix)
    
    def distance(self, x1, x2):
        if self.metric == "euclidean":
            return np.linalg.norm(x1 - x2)
        elif self.metric == "manhattan":
            denominateur = np.linalg.norm(x1) * np.linalg.norm(x2)
            if denominateur == 0:
                return 0
            return 1 - np.dot(x1, x2) / denominateur
        elif self.metric == "cosine":
            return 1 - np.dot(x1, x2) / (np.linalg.norm(x1) * np.linalg.norm(x2))
        elif self.metric == "minkowski":
            return np.sum(np.abs(x1 - x2) ** self.k) ** (1 / self.k)
        elif self.metric == "hamming":
            return np.sum(x1 != x2) / len(x1)
        elif self.metric == "jaccard":
            intersection = np.sum(np.logical_and(x1, x2))
            union = np.sum(np.logical_or(x1, x2))
            if union == 0:
                return 0
            return 1 - intersection / union
        elif self.metric == "mahalanobis":
            cov_matrix = np.cov(self.x_train, rowvar=False)
            inv_cov_matrix = np.linalg.inv(cov_matrix)
            diff = x1 - x2
            return np.sqrt(np.dot(np.dot(diff.T, inv_cov_matrix), diff))
        else:
            raise ValueError(f"Metrique inconnue : {self.metric}.\n Les métriques prises en charge sont : 'euclidean', 'manhattan', 'cosine', 'minkowski', 'hamming', 'jaccard', 'mahalanobis' ")
        
    def predict(self, X_pred):

        list_distance = []
        d = 0
        for i in range(self.x_train.shape[0]): 
            d = self.distance(X_pred, self.x_train[i])
            list_distance.append(d)
        print(list_distance)
        idx = np.argsort(list_distance)[:self.k]
        # list_triee = sorted(list_distance)
        # dist_dec = list_triee[:self.k]
        # idx = []

        # for i in dist_dec :
        #     idx.append(list_distance.index(i))
        if self.pb =='reg':
            return np.mean(self.y_train[idx])
        elif self.pb == "class":
            return Counter(self.y_train[idx]).most_common(1)[0][0]
```

```python

Height = [6,6.11,5.9,5.8,5.3,5.6,5.5]
Age = [40,26,30,32,33,34,35]
Weight = [60,55,56,78,75,78,80]
X_train = np.array(list(zip(Height,Age)))
y_train = np.array(Weight)
```

```python
%%time
knn = KNN(k=3, metric="euclidean", pb="reg")

knn.fit(X_train, y_train)

prediction = knn.predict([5.8, 37])
prediction
```
